```
import requests
import json
from datetime import datetime

base_url = 'https://gitlab.com/api/v4/projects/{project_id}/repository/tree'
contributors_url = 'https://gitlab.com/api/v4/projects/{project_id}/repository/contributors'
project_details_url = 'https://gitlab.com/api/v4/projects/{project_id}'

project_id = '1794617'
access_token = 'glpat-aKrZxxiDFhZ2_42Pnscu'

params = {
    'private_token': access_token,
    'per_page': 100,
    'recursive': True,
}

def get_files_from_all_pages(url):
    all_files = []
    page = 1
    while True:
        response = requests.get(url.format(project_id=project_id), params={**params, 'page': page})
        if response.status_code == 200:
            data = response.json()
            all_files.extend(data)
            if len(data) < 100:  # Assuming 100 items per page, adjust as needed
                break
            page += 1
        else:
            print('Failed to retrieve data:', response.status_code, response.text)
            return None
    return all_files

# Get files data
files_data = get_files_from_all_pages(base_url)

if files_data is not None:
    try:
        contributors_data = requests.get(contributors_url.format(project_id=project_id), params=params)
        project_data = requests.get(project_details_url.format(project_id=project_id), params=params)

        if contributors_data.status_code == 200 and project_data.status_code == 200:
            contributors_info = [{"name": contributor["name"], "email": contributor["email"]} for contributor in contributors_data.json()]
            project_name = project_data.json()["name"]

            output = {
                "id": project_id,
                "name": project_name,
                "contributors": contributors_info,
                "contents": {
                    "meta": {
                        "folders_count": 0,
                        "files_count": 0
                    },
                    "folders": [],
                    "files": []
                }
            }

            for item in files_data:
                if item['type'] == 'tree':
                    output["contents"]["meta"]["folders_count"] += 1
                    output["contents"]["folders"].append({
                        "folder_name": item["name"],
                        "folder_path": item["path"],
                        "folder_mode": item["mode"]
                    })
                elif item['type'] == 'blob':
                    output["contents"]["meta"]["files_count"] += 1
                    output["contents"]["files"].append({
                        "file_name": item["name"],
                        "file_path": item["path"],
                        "file_mode": item["mode"]
                    })

            timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
            file_name = f"repo-query-{timestamp}.json"
            print(json.dumps(output, indent=4))
            with open(file_name, 'w') as file:
                json.dump(output, file, indent=4)
        else:
            print('Failed to retrieve contributors or project data')

    except json.JSONDecodeError as e:
        print('Error decoding JSON response:', e)
```
