```
gender_crosswalk = spark.read.format('csv') \
    .option('header',True)\
    .option('inferschema',True)\
    .option('delimiter','\t')\
    .load('dbfs:/FileStore/tables/us_mock_data/gender_crosswalk.txt')
display(gender_crosswalk)
```

```
group_crosswalk = spark.read.format('csv') \
    .option('header',True)\
    .option('inferschema',True)\
    .option('delimiter','\t')\
    .load('dbfs:/FileStore/tables/us_mock_data/group_crosswalk.txt')
display(group_crosswalk)
```

###renamed columns to remove ambiguity

```
relation_crosswalk = spark.read.format('csv') \
    .option('header',True)\
    .option('inferschema',True)\
    .option('delimiter','\t')\
    .load('dbfs:/FileStore/tables/us_mock_data/relation_crosswalk.txt')
display(relation_crosswalk)

relation_crosswalk = relation_crosswalk.withColumnRenamed('Rollup_Code','relation_code')\
    .withColumnRenamed('Rollup_Description','relation_description')
display(relation_crosswalk)
```

#Joining Tables

## sample_demo( join demographic_mock_data with gender_crosswalk)

```
sample_demo = demographic_mock_data.alias('demo').join(
    gender_crosswalk.alias('gender'),
    f.col("demo.gender")==f.col("gender.code"),
    "left"
    )
display(sample_demo)
```

##sample_demo(join sample_demo with relation_crosswalk)

```
sample_demo =sample_demo.join(
    relation_crosswalk,
    sample_demo.RELATIONSHIP==relation_crosswalk.relation_code,'left'
    ).filter(~((f.col('Rollup_Code')== 'F') & (f.col('relation_description')== 'Father')))
display(sample_demo)
```

##sample_enroll(join enrollment_mock_data with group_crosswalk)

```
sample_enroll = enrollment_mock_data.alias('enroll').join(group_crosswalk,
                                 on="GROUP_ID", how='left')
display(sample_enroll)
```

##sample_enroll(join enrollment_mock_data with coverage_crosswalk)

```
sample_enroll = sample_enroll.join(coverage_crosswalk,
                                   sample_enroll.COVERAGE_TYPE == coverage_crosswalk.Coverage_ID,
                                   'left')
display(sample_enroll)
```

##sample_enroll(join enrollment_mock_data with plan_crosswalk)

```
sample_enroll =sample_enroll.join(plan_crosswalk,
                                  on="PLAN_ID",
                                  how='left')

display(sample_enroll)
```

##final(join sample_demo and sample_enroll)

```
final = sample_demo.join(sample_enroll,
                         on="MEMBER_ID",
                         how='left')

display(final)
```

```
target = final.select(
    f.row_number().over(Window.orderBy(f.monotonically_increasing_id())).alias("Abacus_Record_ID"),
    f.expr("substr(EMPLOYEE_ID,-3,3)||row_number() OVER (PARTITION BY EMPLOYEE_ID ORDER BY 1)||GENDER||substr(date(DOB),-2,2)").alias("Abacus_Member_ID"),
    'MEMBER_ID',    
    f.col("EMPLOYEE_ID").alias("Subscriber_ID"),
    f.col("FIRST_NAME").alias("Member_First_Name"),
    f.col("LAST_NAME").alias("Member_Last_Name"),
    f.col("MIDDLE_NAME").alias("Member_Middle_Name"),
    f.lit(None).alias("Member_Prefix_Name"),
    f.lit(None).alias("Member_Suffix_Name"),
    f.col("Rollup_Description").alias("Member_Gender"),
    f.date_format(f.col("DOB"),"MM/dd/yyyy").alias("Member_Date_of_Birth"),
    f.col("relation_code").alias("Member_Relationship_Code"),
    f.col("relation_description").alias("Member_Relationship_Description"),
    f.col("PERSON_CODE").alias("Member_Person_Code"),
    f.col("ADDRESS_1").alias("Member_Address_Line_1"),
    f.col("ADDRESS_2").alias("Member_Address_Line_2"),
    f.col("CITY").alias("Member_City"),
    f.col("STATE").alias("Member_State"),
    f.col("COUNTY").alias("Member_County"),
    f.col("ZIP").alias("Member_Postal_Code"),
    f.lit("US").alias("Member_Country"),
    f.lit(None).alias("Member_Home_Phone"),
    f.lit(None).alias("Member_Work_Phone"),
    f.lit(None).alias("Member_Mobile_Phone"),
    f.lit(None).alias("Member_Email"),
    f.lit(None).alias("Member_Is_Deceased"),
    f.lit(None).alias("Member_Date_of_Death"),
    f.lit(None).alias("Member_Deceased_Reason"),
    f.col("GROUP_ID").alias("Enrollment_Group_ID"),
    f.col("GROUP_NAME").alias("Enrollment_Group_Name"),
    f.lit(None).alias("Enrollment_SubGroup_ID"),
    f.lit(None).alias("Enrollment_SubGroup_Name"),
    f.col("Coverage_ID").alias("Enrollment_Coverage_Code"),
    f.col("Coverage_Description").alias("Enrollment_Coverage_Description"),
    f.col("PLAN_ID").alias("Enrollment_Plan_ID"),
    f.col("Plan Name").alias("Enrollment_Plan_Name"),
    f.col("BENEFIT_TYPE").alias("Enrollment_Plan_Coverage"),
    f.when(f.col("BENEFIT_TYPE").contains("Medical"),f.date_format(f.col("enroll.EFFECTIVE_DATE"),"MM/dd/yyyy")).alias("Enrollment_Medical_Effective_Date"),
    f.when(f.col("BENEFIT_TYPE").contains("Medical"),f.date_format(f.col("enroll.TERMINATION_DATE"),"MM/dd/yyyy")).alias("Enrollment_Medical_Termination_Date"),
     f.when(f.col("BENEFIT_TYPE").contains("Dental"),f.date_format(f.col("enroll.EFFECTIVE_DATE"),"MM/dd/yyyy")).alias("Enrollment_Dental_Effective_Date"),
    f.when(f.col("BENEFIT_TYPE").contains("Dental"),f.date_format(f.col("enroll.TERMINATION_DATE"),"MM/dd/yyyy")).alias("Enrollment_Dental_Termination_Date"),
     f.when(f.col("BENEFIT_TYPE").contains("Vision"),f.date_format(f.col("enroll.EFFECTIVE_DATE"),"MM/dd/yyyy")).alias("Enrollment_Vision_Effective_Date"),
    f.when(f.col("BENEFIT_TYPE").contains("Vision"),f.date_format(f.col("enroll.TERMINATION_DATE"),"MM/dd/yyyy")).alias("Enrollment_Vision_Termination_Date"),
    f.col("VENDOR").alias("Enrollment_Vendor_Name"),
    f.lit("Menber Enrollment").alias("Souce_File_Name"),
    f.date_format(f.current_date(),"MM/dd/yyyy").alias("File_Ingestion_Date")

    

)
display(target)
```
